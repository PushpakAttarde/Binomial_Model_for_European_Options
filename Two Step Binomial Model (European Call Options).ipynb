```python
#Two Step Binomial Model_2
#European Call Option
# S = Current Stock Price, K  = Strike Price, u = Up Factor, d = Down Factor, r = risk free rate, T = Time to expiry
```

```python
import numpy as np
from tabulate import tabulate
S = float(input("Enter Current Stock Price: "))
K = float(input("Enter Strike Price: "))
u = float(input("Enter Up Factor: "))
d = float(input("Enter Down Factor: "))
r = float(input("Enter Risk free rate: "))
T = float(input("Enter Time Interval: ") )

S_u = S * u
S_d = S * d
S_uu = S * u ** 2
S_dd = S * d ** 2
S_ud = S * u * d
C_uu = max(0, S_uu - K)
C_dd = max(0, S_dd - K)
C_ud = max(0, S_ud - K)

p = (np.exp(r * T) - d) / (u - d)

C_u = np.exp(-r * T) * (p * C_uu + (1-p) * C_ud)
C_d = np.exp(-r * T) * (p * C_ud + (1-p) * C_dd)
C = np.exp(-r * T) * (p * C_u + (1-p) * C_d)

table_data = [
        ["p (Risk-neutral probability)", f"{p:.4f}"],
        ["C_uu (Call price at up-up node)", f"{C_uu:.4f}"],
        ["C_ud (Call price at up-down node)", f"{C_ud:.4f}"],
        ["C_dd (Call price at down-down node)", f"{C_dd:.4f}"],
        ["C_u (Call price at up node)", f"{C_u:.4f}"],
        ["C_d (Call price at down node)", f"{C_d:.4f}"],
        ["C (Call price at initial node)", f"{C:.4f}"]
    ]

print(tabulate(table_data, headers=["Description", "Value"], tablefmt="grid"))
```

```python
import numpy as np
from tabulate import tabulate
def binomial_call_option_prices(S, K, u, d, r, T):
  S_u = S * u
  S_d = S * d
  S_uu = S * u ** 2
  S_dd = S * d ** 2
  S_ud = S * u * d
  C_uu = max(0, S_uu - K)
  C_dd = max(0, S_dd - K)
  C_ud = max(0, S_ud - K)

  p = (np.exp(r * T) - d) / (u - d)

  C_u = np.exp(-r * T) * (p * C_uu + (1-p) * C_ud)
  C_d = np.exp(-r * T) * (p * C_ud + (1-p) * C_dd)
  C = np.exp(-r * T) * (p * C_u + (1-p) * C_d)

  table_data = [
        ["p (Risk-neutral probability)", f"{p:.4f}"],
        ["C_uu (Call price at up-up node)", f"{C_uu:.4f}"],
        ["C_ud (Call price at up-down node)", f"{C_ud:.4f}"],
        ["C_dd (Call price at down-down node)", f"{C_dd:.4f}"],
        ["C_u (Call price at up node)", f"{C_u:.4f}"],
        ["C_d (Call price at down node)", f"{C_d:.4f}"],
        ["C (Call price at initial node)", f"{C:.4f}"]
    ]

  print(tabulate(table_data, headers=["Description", "Value"], tablefmt="grid"))
```
