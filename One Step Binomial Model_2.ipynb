```python
#One Step Binomial Model_2
#European Call Option
# S = Current Stock Price, K  = Strike Price, u = Up Factor, d = Down Factor, r = risk free rate, T = Time to expiry
```

```python
import numpy as np

def call_option_price(S, K, u, d, r, T):
  S_u = S * u
  S_d = S * d
  C_u = max(0, S_u - K)
  C_d = max(0, S_d - K)

  p = (np.exp(r * T) - d) / (u - d)

  call_option_price = np.exp(-r * T) * (p * C_u + (1-p) * C_d)
  return call_option_price
```

```python
call_price = call_option_price(20, 21, 1.1, 0.9, 0.12, 0.25)
print(call_price)
```
